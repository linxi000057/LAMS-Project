```python
import os
import csv
import pandas as pd
```

```python
data_file = "../Top10_Conditions_by_Year.csv"
```

```python
data_file_pd = pd.read_csv(data_file)
data_file_pd.head(2)
```

```python
pivoted_table = data_file_pd.pivot(index="Year", columns="Condition", values="Count")
pivoted_table.reset_index(inplace=True)
pivoted_table
```

```python
pivoted_table.to_csv("Top10_Pivoted.csv", index=False)
```
