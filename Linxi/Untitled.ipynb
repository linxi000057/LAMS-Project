```python
import requests
url='https://gismaps.vita.virginia.gov/arcgis/rest/services/VA_Base_layers/VA_Admin_Boundaries/MapServer/1/query?where=1%3D1&outFields=*&outSR=4326&f=json'
r = requests.get(url)
data = r.json()
```

```python
data.keys()
```

```python
data['features'][0]['geometry']
```

```python
r = requests.get('http://127.0.0.1:5000/data')
app_data = r.json()
counties = set([i.get('County') for i in app_data])
counties
```

```python
features = data['features']
other_counties = set([i.get('attributes').get('NAME') for i in features])
other_counties
```

```python
counties_with_city = list(counties - other_counties)
counties_without_city = [i.replace(" (city)","") for i in counties_with_city]
set(counties_without_city) - other_counties
```

# Combine Data

```python
for row in app_data:
    county = row.get('County').replace(" (city)","")
    matches = [i.get('geometry') for i in features if i.get('attributes').get('NAME') == county]
    row.update({'geometry': matches[0]})
```
